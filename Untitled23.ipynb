```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import sys
import cv2

def load_pretrained_model():
    """Load a pre-trained MNIST model from Keras"""
    mnist = keras.datasets.mnist
    (x_train, y_train), (x_test, y_test) = mnist.load_data()

    x_train = x_train / 255.0
    x_test = x_test / 255.0

    try:
        model = keras.models.load_model('mnist_model.h5')
        print("Loaded existing model")
    except:
        print("Training new model...")
        model = keras.Sequential([
            keras.layers.Flatten(input_shape=(28, 28)),
            keras.layers.Dense(128, activation='relu'),
            keras.layers.Dropout(0.2),
            keras.layers.Dense(10, activation='softmax')
        ])

        model.compile(optimizer='adam',
                     loss='sparse_categorical_crossentropy',
                     metrics=['accuracy'])

        model.fit(x_train, y_train, epochs=5, verbose=1)
        model.save('mnist_model.h5')
        print(f"Model trained with accuracy: {model.evaluate(x_test, y_test, verbose=0)[1]:.4f}")

    return model

def segment_digits(img_array):
    """Segment individual digits from the image"""
    # Normalize to 0-255 range for OpenCV
    if img_array.max() <= 1.0:
        img_array = (img_array * 255).astype(np.uint8)
    else:
        img_array = img_array.astype(np.uint8)

    # Threshold the image
    _, binary = cv2.threshold(img_array, 127, 255, cv2.THRESH_BINARY_INV)

    # Find contours
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Get bounding boxes and sort left to right
    digit_boxes = []
    for contour in contours:
        x, y, w, h = cv2.boundingRect(contour)
        # Filter out noise (too small regions)
        if w > 5 and h > 5:
            digit_boxes.append((x, y, w, h))

    # Sort by x-coordinate (left to right)
    digit_boxes.sort(key=lambda box: box[0])

    # Extract individual digits
    digits = []
    for x, y, w, h in digit_boxes:
        digit_img = img_array[y:y+h, x:x+w]
        digits.append(digit_img)

    return digits, digit_boxes

def preprocess_digit(digit_img):
    """Preprocess a single digit for prediction"""
    # Normalize
    if digit_img.max() > 1.0:
        digit_img = digit_img / 255.0

    # Add padding to make it square
    h, w = digit_img.shape
    size = max(h, w)
    padded = np.zeros((size, size))
    y_offset = (size - h) // 2
    x_offset = (size - w) // 2
    padded[y_offset:y_offset+h, x_offset:x_offset+w] = digit_img

    # Resize to 28x28
    resized = cv2.resize(padded, (28, 28))

    # Add batch dimension
    resized = np.expand_dims(resized, axis=0)

    return resized

def recognize_number(npy_file_path, model):
    """Load .npy file and recognize the complete number"""
    # Load the numpy array
    img_array = np.load(npy_file_path)

    print(f"Loaded image with shape: {img_array.shape}")
    print(f"Value range: [{img_array.min():.2f}, {img_array.max():.2f}]")

    # Handle multi-dimensional arrays
    if len(img_array.shape) > 2:
        img_array = img_array.squeeze()
        if len(img_array.shape) > 2:
            # Take first channel if still multi-channel
            img_array = img_array[:, :, 0]

    # Segment individual digits
    print("\nSegmenting digits...")
    digits, boxes = segment_digits(img_array)

    if len(digits) == 0:
        print("No digits found in the image!")
        return None, []

    print(f"Found {len(digits)} digit(s)")

    # Recognize each digit
    recognized_digits = []
    confidences = []

    for i, digit_img in enumerate(digits):
        processed = preprocess_digit(digit_img)
        predictions = model.predict(processed, verbose=0)
        predicted_digit = np.argmax(predictions[0])
        confidence = predictions[0][predicted_digit] * 100

        recognized_digits.append(predicted_digit)
        confidences.append(confidence)
        print(f"  Digit {i+1}: {predicted_digit} (confidence: {confidence:.2f}%)")

    # Combine digits into a number
    number = int(''.join(map(str, recognized_digits)))
    avg_confidence = np.mean(confidences)

    return number, confidences
```
